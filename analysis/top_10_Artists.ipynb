```python
import pandas as pd
from pathlib import Path
```

```python
# Load data
file_path = Path("../data/columns_final_df.json")
df = pd.read_json(file_path)
```

```python
df.head()
```

```python
#Count the Total Number of Unique Artists
df['Artist'].nunique()
```

```python
# Find the Top 10 Artists with the most Hot 100 songs
df['Artist'].value_counts().head(11) 
```

```python
#Find Number of One Hit Wonders

#Set Count of each artist into a Variable
artist_counts = df['Artist'].value_counts()

#Filter for Artists who only appear once as the Recording Artist and Count them
one_hit_wonders = artist_counts[artist_counts ==1]
one_hit_wonders.count()
```

```python
#Find Top 10 Artists with Most Primary Features
df['First Feature'].value_counts().head(10)
```

```python
# Find How Many Total Hits Drake Has
drake_df = df.loc[(df["Artist"] == "Drake") | (df["First Feature"] == "Drake") 
| (df["Second Feature"] == "Drake"), :]
drake_df
```

```python
#Find Total Hits for Drake
total_hits = drake_df.shape[0]
total_hits
```
