```python
# Define start and end years
start_year = 2006
end_year = 2024

# Initialize a dictionary to store DataFrames by year
dataframes_by_year = {}

# Initialize an empty list to store all data
data = []

# Initialize a dictionary to store the title counts for each year
title_counts_by_year = {}

# Loop through each year and fetch the year-end chart data
for year in range(start_year, end_year + 1):
    print(f"Fetching year-end data for {year}...")
    try:
        # Initialize a list to store the current year's data
        year_data = []

        # Fetch the year-end chart for the given year
        chart = billboard.ChartData('hot-100-songs', year=year)
        for entry in chart:
            entry_data = {
                "Year": year,
                "Rank": entry.rank,
                "Title": entry.title,
                "Artist": entry.artist
            }
            data.append(entry_data)  # Add to overall data list
            year_data.append(entry_data)  # Add to year-specific list

        # Create a DataFrame for the current year
        year_df = pd.DataFrame(year_data)
        dataframes_by_year[year] = year_df

        # Count the number of titles for the current year
        title_counts_by_year[year] = len(year_df)

    except Exception as e:
        print(f"Error fetching data for {year}: {e}")

print("Year-end data fetching complete!")

# Display the total title counts for each year. \n is an escape sequence that inserts a newline character, creating a blank line before the message.
print("\nTitle counts by year:")
for year, count in title_counts_by_year.items():
    print(f"{year}: {count} titles")
```

```python
def get_token():
    auth_string = client_id + ":" + client_secret
    auth_bytes = auth_string.encode("utf-8")
    auth_base64 = str(base64.b64encode(auth_bytes), "utf-8")

    url = "https://accounts.spotify.com/api/token"
    headers = {
        "Authorization": "Basic " + auth_base64,
        "Content-Type": "application/x-www-form-urlencoded"
    }
    data = {"grant_type": "client_credentials"}
    result = post(url, headers=headers, data=data)
    json_result = json.loads(result.content)
    token = json_result["access_token"]
    return token
```

```python
def get_auth_header(token):
    return {"Authorization": "Bearer " + token}
```

```python
def search_all_songs(token, titles_and_artists):
    results = []
    for song_name, artist_name in titles_and_artists:
        print(f"Searching for: '{song_name}' by '{artist_name}'")
        result = search_for_song(token, song_name, artist_name)
        if result:
            if result['tracks']['items']:
                track = result['tracks']['items'][0]
                artist_id = track['artists'][0]['id']
                duration_ms = track['duration_ms']
                popularity = track['popularity']
                release_date = track['album']['release_date']
                
                # Get the artist's genres using the artist's Spotify ID
                artist_url = f"https://api.spotify.com/v1/artists/{artist_id}"
                artist_response = requests.get(artist_url, headers=get_auth_header(token))
                
                genres = []
                if artist_response.status_code == 200:
                    artist_data = artist_response.json()
                    genres = artist_data.get('genres', [])

                # Convert ms to minutes and seconds format
                minutes, seconds = divmod(duration_ms // 1000, 60)
                duration_formatted = f"{minutes}:{seconds:02d}"
                
                results.append({
                    "Song Name": track['name'],
                    "Artist": ", ".join(artist['name'] for artist in track['artists']),
                    "Spotify ID": track['id'],
                    "Genres": ", ".join(genres),
                    "Duration": duration_formatted,
                    "Popularity": popularity,
                    "Release Date": release_date
                })
            else:
                results.append({"Song Name": song_name, "Artist": artist_name, "Error": "Track not found"})
        else:
            results.append({"Song Name": song_name, "Artist": artist_name, "Error": "No response from Spotify"})
        
        # Add a small delay between requests to prevent rate limiting
        time.sleep(1)

    # Sort the results by release date
    sorted_results = sorted(results, key=lambda x: x.get("Release Date", "9999-99-99"))
    
    return results
```

```python
def search_for_song(token, song_name, artist_name):
    url = "https://api.spotify.com/v1/search"
    headers = get_auth_header(token)
    params = {
        "q": f"track:{song_name} artist:{artist_name}",
        "type": "track",
        "limit": 1
    }
    
    # Perform the request
    response = requests.get(url, headers=headers, params=params)
    
    # Parse the JSON result
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error for song '{song_name}' by '{artist_name}': {response.status_code}, {response.text}")
        return None

# Create the titles_and_artists list from the DataFrame
titles_and_artists = list(zip(df_all_years['Title'], df_all_years['Artist']))

# Get the token by calling the get_token() function
token = get_token()
# Now, use the titles_and_artists list for searching songs
access_token = token  # Your obtained access token
spotify_results = search_all_songs(access_token, titles_and_artists)

# Convert the results to a more readable format
spotify_results_json = json.dumps(spotify_results, indent=4)
print(spotify_results_json)
```

```python
def search_missing_songs_by_name_and_artist(token, missing_songs):
    url = "https://api.spotify.com/v1/search"
    headers = get_auth_header(token)
    results = []

    for song_name, artist_name in missing_songs:
        # First, try searching with both song name and artist
        params = {
            "q": f"track:{song_name} artist:{artist_name}" if artist_name else f"track:{song_name}",
            "type": "track",
            "limit": 1
        }

        response = requests.get(url, headers=headers, params=params)
        if response.status_code == 200:
            data = response.json()
            if data["tracks"]["items"]:
                track = data["tracks"]["items"][0]
                results.append({
                    "Song Name": song_name,
                    "Artist": ", ".join([artist["name"] for artist in track["artists"]]),
                    "Spotify ID": track["id"],
                    "Genres": ", ".join(track.get("album", {}).get("genres", [])),
                    "Duration": f"{track['duration_ms'] // 60000}:{(track['duration_ms'] // 1000) % 60:02}",
                    "Popularity": track["popularity"],
                    "Release Date": track["album"]["release_date"],
                    "Error": None
                })
            else:
                # Fallback to search by only song name
                fallback_params = {
                    "q": f"track:{song_name}",
                    "type": "track",
                    "limit": 1
                }
                fallback_response = requests.get(url, headers=headers, params=fallback_params)
                if fallback_response.status_code == 200:
                    fallback_data = fallback_response.json()
                    if fallback_data["tracks"]["items"]:
                        fallback_track = fallback_data["tracks"]["items"][0]
                        results.append({
                            "Song Name": song_name,
                            "Artist": ", ".join([artist["name"] for artist in fallback_track["artists"]]),
                            "Spotify ID": fallback_track["id"],
                            "Genres": ", ".join(fallback_track.get("album", {}).get("genres", [])),
                            "Duration": f"{fallback_track['duration_ms'] // 60000}:{(fallback_track['duration_ms'] // 1000) % 60:02}",
                            "Popularity": fallback_track["popularity"],
                            "Release Date": fallback_track["album"]["release_date"],
                            "Error": None
                        })
                    else:
                        results.append({
                            "Song Name": song_name,
                            "Artist": artist_name,
                            "Spotify ID": None,
                            "Genres": None,
                            "Duration": None,
                            "Popularity": None,
                            "Release Date": None,
                            "Error": "Track not found"
                        })
                else:
                    results.append({
                        "Song Name": song_name,
                        "Artist": artist_name,
                        "Spotify ID": None,
                        "Genres": None,
                        "Duration": None,
                        "Popularity": None,
                        "Release Date": None,
                        "Error": f"API error: {fallback_response.status_code}"
                    })
        else:
            print(f"Error for song '{song_name}' by '{artist_name}': {response.status_code}, {response.text}")
            results.append({
                "Song Name": song_name,
                "Artist": artist_name,
                "Spotify ID": None,
                "Genres": None,
                "Duration": None,
                "Popularity": None,
                "Release Date": None,
                "Error": f"API error: {response.status_code}"
            })

    return results

# Prepare the list of missing songs with both song name and artist
missing_songs = [
    ("For You I Will (Confidence)", "Teddy Geiger"),
    ("Ransom", "Lil Tecca")
]

# Call the function with the access token and missing songs
missing_song_results = search_missing_songs_by_name_and_artist(access_token, missing_songs)

# Convert the results to a DataFrame
missing_songs_df = pd.DataFrame(missing_song_results)
print(missing_songs_df)
```
